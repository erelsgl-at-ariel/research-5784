 ## Sorting

```python
df.sort_values(by=['GDP per capita'],ascending=False)[['Country or region', 'GDP per capita']]
```

```python
def to_lower_case(row):
    return row['Country or region'].lower()  # return only the country, and in lower-case

df_lower = df.apply(to_lower_case, axis=1)  # Apply the function row-by-row
df_lower
```

```python
df['Country or region'] = df_lower   # update the original DataFrame
df.head(10)
```

```python
df['Country or region Uppercase'] = df.apply(lambda row: row['Country or region'].upper(), axis=1)
df
```
