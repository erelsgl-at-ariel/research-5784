```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line plot

```python
plt.plot([1,2,3,4,5], [1,4,9,16,25]);    # the semicolon hides the output text (only the figure is shown)
```

```python
x = np.linspace(0,20, 101)  # returns an array with 101 linearly-spaced numbers between 0 and 20.
plt.plot(x, np.sin(x));     # the semicolon hides the output text (only the figure is shown)
```

```python
# Adding labels:

plt.plot(x, np.sin(x))
plt.xlabel("input")
plt.ylabel("output")
plt.title("sinus")
plt.grid();
```

```python
# Several plots on the same axes:

plt.plot(x, np.sin(x), 'r-', label="sin(x)")  # red, solid line
plt.plot(x, np.cos(x/2), 'b.-', label="cos(x/2)")   # blue, dotted line
plt.xlabel("input")
plt.ylabel("output")
plt.title("trigonometric functions")
plt.legend()
plt.grid();
```

## Scatter plot

```python
NUM_SUBJECTS=100
ages =   np.random.randint(20, 60, size=NUM_SUBJECTS)
grades = np.random.normal(ages*2, 10, size=NUM_SUBJECTS)
plt.scatter(ages, grades);
```

```python
colors = [grades > ages*2]
print(colors)
plt.scatter(ages, grades, c = colors)
plt.show()
```

```python
print('~~~ normal distribution: ~~~')
normal_dist = np.random.normal(10, 1, size=100000)
plt.hist(normal_dist, bins = 100)
plt.show()
```

```python
binomial_dis = np.random.binomial(n=10, p=0.5, size=10000)
print('~~~ binomial distribution: ~~~')
plt.hist(binomial_dis, bins = 100)
plt.show()
```

```python
uniform_dis = np.random.uniform(size=100000)
print('~~~ uniform distribution: ~~~')
plt.hist(uniform_dis, bins = 100)
plt.show()
```

```python
poisson_dis = np.random.poisson(lam=2, size=10000)
print('~~~ poisson distribution: ~~~')
plt.hist(poisson_dis, bins = 100)
plt.show()
```

```python
random_array = np.random.random(10000)
plt.hist(random_array, bins = 100)
plt.show()
```

## Subplots

```python
SIZE=10000
BINS=100
fig, axes = plt.subplots(2, 3, sharex=True, sharey=True)
axes[0,0].hist(np.random.normal(0, 1, size=SIZE), bins=BINS)
axes[0,1].hist(np.random.normal(0, 2, size=SIZE), bins=BINS)
axes[0,2].hist(np.random.normal(0, 3, size=SIZE), bins=BINS)
axes[1,0].hist(np.random.normal(10, 1, size=SIZE), bins=BINS)
axes[1,1].hist(np.random.normal(10, 2, size=SIZE), bins=BINS)
axes[1,2].hist(np.random.normal(10, 3, size=SIZE), bins=BINS)
plt.show()
```

## Plotting 3-dimensional functions

```python
def g(xx,yy):
    return xx**3 + yy**2

x=np.linspace(-6, 6, 31)
y=np.linspace(-6, 6, 31)
xx, yy = np.meshgrid(x,y)
print("xx", xx[0:3,:])
print("yy", yy[0:3,:])
z = g(xx, yy)
print("z", z[0:3,:])
```

```python
plt.contourf(x, y, z)
plt.axis('scaled')
plt.colorbar();  # add color legend at the right
```

## Saving plots to a file

```python
plt.contourf(x, y, z)
plt.axis('scaled')
plt.colorbar()
plt.savefig("image.png")
```
